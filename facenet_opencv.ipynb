```python
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
from imutils.video import VideoStream
import numpy as np
import imutils
import time
import cv2
import os

m2 = load_model("vgg16.model")
m1=load_model("mask_detector.model")
def dpm(frm, faceNet, m1,m2):

    (h, w) = frm.shape[:2]
    blb = cv2.dnn.blobFromImage(frm, 1.0, (224, 224),(104.0, 177.0, 123.0))

    faceNet.setInput(blb)
    det = faceNet.forward()
    print(det.shape)
   
    faces = []
    locs = []
    preds = []


    for i in range(0, det.shape[2]):
        confidence = det[0, 0, i, 2]


        if confidence > 0.5:
            bx = det[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x_start, y_start, x_end, y_end) = bx.astype("int")
            (x_start, y_start) = (max(0, x_start), max(0, y_start))
            (x_end, y_end) = (min(w - 1, x_end), min(h - 1, y_end))
            fc = frm[y_start:y_end, x_start:x_end]
            fc = cv2.cvtColor(fc, cv2.COLOR_BGR2RGB)
            fc = cv2.resize(fc, (224, 224))
            fc = img_to_array(fc)
            fc = preprocess_input(fc)
            faces.append(fc)
            locs.append((x_start, y_start, x_end, y_end))

    if len(faces) > 0:
        faces = np.array(faces, dtype="float32")
        l1=m1.predict(faces,batch_size=32)
        l2=m2.predict(faces,batch_size=32)
        l3=[]
        l3.append(l1[0][0]+l2[0][0])
        l3.append(l1[0][1]+l2[0][1])
        preds.append(l3)
    return (locs, preds)

ptxtpath= r"face_detector\deploy.prototxt"
wpath = r"face_detector\res10_300x300_ssd_iter_140000.caffemodel"
faceNet = cv2.dnn.readNet(ptxtpath, wpath)


vs = VideoStream(src=0).start()


while True:
    frm = vs.read()
    frm = imutils.resize(frm, width=400)
    (locs, preds) = dpm(frm, faceNet, m1,m2)
   
    for (bx, pred) in zip(locs, preds):
        (x_start, y_start, x_end, y_end) = bx
        (mask, withoutMask) = pred
        label = "Mask" if mask > withoutMask else "No Mask"
        color = (0, 255, 0) if label == "Mask" else (0, 0, 255)
        num=max(mask,withoutMask)
        num=num/2
        label = "{}: {:.2f}%".format(label, num * 100)
        cv2.putText(frm, label, (x_start, y_start - 10),
            cv2.FONT_HERSHEY_SIMPLEX, 0.45, color, 2)
        cv2.rectangle(frm, (x_start, y_start), (x_end, y_end), color, 2)

    cv2.imshow("Frame", frm)
    key = cv2.waitKey(1) & 0xFF
   
    if key == ord("q"):
        break


cv2.destroyAllWindows()
vs.stop()
```
